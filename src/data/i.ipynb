```python
import pandas as pd

# Ruta del archivo CSV
csv_file_path = './Data.csv'

# Leer el archivo CSV utilizando pandas
data = pd.read_csv(csv_file_path)

# Convertir la columna 'timestamp' en formato de fecha y hora
data['timestamp'] = pd.to_datetime(data['timestamp'])

# Obtener el a単o y el mes por separado en columnas adicionales
data['year'] = data['timestamp'].dt.year
data['month'] = data['timestamp'].dt.month

# Agrupar los datos por a単o y mes y contar las incidencias en cada mes
data_grouped = data.groupby(['year', 'month'])['ID'].count().reset_index()

# Imprimir el resultado
current_year = None
for _, row in data_grouped.iterrows():
    year = row['year']
    month = row['month']
    count = row['ID']

    if year != current_year:
        print(year)
        current_year = year

    print(f"{{ x: {month}, y: {count} }}")
```
